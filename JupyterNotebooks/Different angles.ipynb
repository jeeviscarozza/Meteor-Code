```python
#Turning this into a function
def distance_at_angle(v_initial,t_initial,x_initial,y_initial):
    '''this is used to find the distances the ball lands at given initial conditions, at each angle'''
    x = x_initial
    #first I am making a list of angles
    anglelist = []
    angle = 0
    while angle < math.pi/2:
        angle = angle + math.pi/8
        anglelist.append(angle)
    #now I have a list of angles that I can pick from, now make list of x values
    xlist = []
    #now write a loop for finding the x values at the different angles
    for angle in anglelist:
    #find the time
        t = solving_for_t(angle,y_initial,v_initial)  
    #now use the time to find the x value at that angle. this is still within the loop
        x = v_initial*math.cos(angle)*t
        xlist.append(x)
    return [anglelist,xlist]
```

```python
def find_max_in_list(mylist):
    '''this function finds the maximum value in a list'''
    if len(mylist) == 1:
        higher_number = mylist[0]
        return(higher_number)
    else:
        index=1
        if mylist[1] > mylist[0]:
            higher_number = mylist[1]
        else:
            higher_number = mylist[0]
        while index < len(mylist)-1:
            index = index + 1
            if mylist[index] > higher_number:
                higher_number = mylist[index]
            else:
                higher_number = higher_number   
        return higher_number
```

```python
def distances_at_height(height):
    '''funtion to find the distances the ball lands at when launched from a certain height, over all the angles'''
    temp = distance_at_angle(10,0,0,height)
#with an initial velocity of 10, at a height we plug in, we can find the distance it lands at each angle
    just_x_list = temp[1]
    return just_x_list
```

```python
#this is to find the maximum distance versus height
listofheights = []
y_initial = 10
#making the list of heights. this will be the on the x axis of the graph
while y_initial < 200:
    y_initial = y_initial + 10
    listofheights.append(y_initial)
#now trying to get the maximum distance for each height
maxdistancelist = []
for y_initial in listofheights:
    xlist = distances_at_height(y_initial)
    max_distance= find_max_in_list(xlist)
    maxdistancelist.append(max_distance)
```
